Now, our stack is all ready... it's time to remove a book from the stack... 

```python
class Book_stack:
  def __init__(self, book_name):
    #make a list to save all the books
    self.books_list = []
    #save the book
    self.books_list.append(book_name)
    self.show_success_message(book_name)

  
  def add_book_to_stack(self, book_name):
    self.books_list.append(book_name)
    self.show_success_message(book_name)

  #-----CHECK THIS OUT-----
  def remove_book(self):
    removed_book = self.books_list.pop()
    return removed_book
  #------------------------

  def show_success_message(self, book_name):
    print("Your book", book_name, "has been added to the book stack")
```

Now, let's try to remove some books

```python
#STEP 1 - create a stack
my_book_stack = Book_stack("The horror of operating systems!")

#STEP 2 - add more and more books to your stack
my_book_stack.add_book_to_stack("The crazy numerical methods in programming")
my_book_stack.add_book_to_stack("Successful turtle stories - while it goes Left on your Python IDE")

#STEP 3 - remove some books
removed_book = my_book_stack.remove_book()
print("Your book", removed_book, "has been removed from the stack")
```

#Queues ðŸšŒðŸšŒðŸšŒ



Queues are quite similar to stack... but wait!... what's the difference if both are similar! ðŸ˜µ

The onely difference is that the thing which goes into your data-structure first, comes out i.e. pops out first...

Let's say you are standing at the back of a Cola Vending machine and restocking it... and you put in some cans of Cola from the back side of the vending machine... So, if some customer comes in and presses the cola button, the can you put in the vending machine first, will be given to the customer. 

Let's make a Cola Vending Machine... 

```python
#totally same as stacks

class cola_machine:
  def __init__(self, cola_can):
    #make a list to save all the books
    self.cola_list = []
    #save the book
    self.cola_list.append(cola_can)
    self.show_success_message(cola_can)

  
  def add_can_to_machine(self, cola_can):
    self.cola_list.append(cola_can)
    self.show_success_message(cola_can)

  #-----CHECK THIS OUT-----
  def remove_can(self):
    removed_can = self.cola_list.pop(0)
    return removed_can
  #------------------------

  def show_success_message(self, cola_can):
    print("Your can", cola_can, "has been added to the cola can queue")
```

If you see ðŸ‘€ the ony change made in the queue is the remove function where we always pop(0). 

See that was simple... ðŸ˜Ž

Now, let's try to add and remove cola cans from our cola vending machine

```python
#STEP 1 - create a stack
cola_queue = cola_machine("Cola Can 1")

#STEP 2 - add more and more books to your stack
cola_queue.add_can_to_machine("Cola Can 2")
cola_queue.add_can_to_machine("Cola Can 3")

#STEP 3 - remove some books
removed_can = cola_queue.remove_can()
print(removed_can, "has been removed from the cola machine")
```
